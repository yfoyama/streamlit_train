```python
def get_data(days,tickers):
    df = pd.DataFrame()

    for company in tickers.keys():

        trk = yf.Ticker(tickers[company])
        hist = trk.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name ='Name'
        df = pd.concat([df,hist])

    return df

days = 20
tickers = {'apple':'AAPL',
           'Meta':'Meta',
           'Google':'GOOGL',
           'microsoft':'MSFT',
           'amazon':'AMZN'
           }

df = get_data(days,tickers)
```

```python
import altair

companies = ['apple','Meta']
data = df.loc[companies]

```

```python
data.sort_index
```

```python
data = data.T.reset_index()
```

```python
data.head()
```

```python
data = pd.melt(data, id_vars=['Date'])
```

```python
data
```

```python
data.rename(columns={'value':'Stock Prices(USD)'},inplace=True)

data
```

```python
import altair as alt

ymin, ymax = 450, 550

chart = (
    alt.Chart(data)
    .mark_line(opacity = 0.8,clip=True)
    .encode(
        x = "Date:T",
        y = alt.Y("Stock Prices(USD)",stack = None, scale=alt.Scale(domain=[ymin,ymax])),
        color = 'Name:N'
    )
)
```
